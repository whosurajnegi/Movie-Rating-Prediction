```python
ratings.drop(['user_id','occupation','zipcode','timestamp'],axis=1, inplace=True)
ratings
```

```python
#Data Visualisation
# user age distribution
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(8,6))
ratings.age_range.hist()
plt.title('User Age Distribution')
plt.xlabel('age_range')
plt.ylabel('Number of Users')
plt.show()
```

```python
ratings['genres'].value_counts().head()
```

```python
ratings['age_range'] = ratings['age_range'].replace({'Under 18':16,'25-34':30,'45-49':47,'50-55':52,'35-44':40,'18-24':21,'56+':60})
ratings
```

```python
ratings['gender']= ratings['gender'].replace({'Female': 0, 'Male':1})
ratings
```

```python
#NLP
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
import re
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
stemmer = PorterStemmer()
lem = WordNetLemmatizer()
corpus=[]
```

```python
for i in range(len(ratings)):
    review = re.sub('[^a-zA-Z]', ' ', ratings['genres'][i])
    review = review.lower()
    review = review.split()
    review = [lem.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]  
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=500)
X = tfidf.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
data = ratings.copy()
data
```

```python
#Vectorisation
vector= pd.DataFrame(X)
vector
```

```python
data.drop(['title', 'genres'], axis=1, inplace=True)
data
```

```python
vectorized_data = pd.concat([data, vector], ignore_index=True,axis=1)
vectorized_data
```

```python
vectorized_data.columns
```

```python
x = vectorized_data.drop(3, axis=1)
y = vectorized_data[3]
```

```python
Y = ratings['rating']
Y
```

```python
#Machine learning
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
#Create train and test data set
from sklearn.model_selection import train_test_split
train, test, train_labels, test_labels = train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
# Logistic Regression
logreg = LogisticRegression()
logreg.fit(train, train_labels)
Y_pred = logreg.predict(test)
acc_log = round(logreg.score(train, train_labels) * 100, 2)
acc_log
```

```python
# Support Vector Machines
svc = SVC()
svc.fit(train, train_labels)
Y_pred = svc.predict(test)
acc_svc = round(svc.score(train, train_labels) * 100, 2)
acc_svc
```

```python
# K Nearest Neighbors Classifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(train, train_labels)
Y_pred = knn.predict(test)
acc_knn = round(knn.score(train, train_labels) * 100, 2)
acc_knn
```

```python
# Gaussian Naive Bayes
gaussian = GaussianNB()
gaussian.fit(train, train_labels)
Y_pred = gaussian.predict(test)
acc_gaussian = round(gaussian.score(train, train_labels) * 100, 2)
acc_gaussian
```

```python
# Linear SVC
linear_svc = LinearSVC()
linear_svc.fit(train, train_labels)
Y_pred = linear_svc.predict(test)
acc_linear_svc = round(linear_svc.score(train, train_labels) * 100, 2)
acc_linear_svc
```

```python
# Stochastic Gradient Descent
sgd = SGDClassifier()
sgd.fit(train, train_labels)
Y_pred = sgd.predict(test)
acc_sgd = round(sgd.score(train, train_labels) * 100, 2)
acc_sgd
```

```python
# Decision Tree
decision_tree = DecisionTreeClassifier()
decision_tree.fit(train, train_labels)
Y_pred = decision_tree.predict(test)
acc_decision_tree = round(decision_tree.score(train, train_labels) * 100, 2)
acc_decision_tree
```

```python
# Random Forest
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(train, train_labels)
Y_pred = random_forest.predict(test)
random_forest.score(train, train_labels)
acc_random_forest = round(random_forest.score(train, train_labels) * 100, 2)
acc_random_forest
```

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 
              'Random Forest', 'Naive Bayes', 'Stochastic Gradient Decent', 'Linear SVC', 
              'Decision Tree'],
    'Score': [acc_svc, acc_knn, acc_log, 
              acc_random_forest, acc_gaussian,  
              acc_sgd, acc_linear_svc, acc_decision_tree]})
models.sort_values(by='Score', ascending=False)
```
